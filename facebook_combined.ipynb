```python
import gzip
import networkx as nx
import matplotlib.pyplot as plt

# Load the graph data from the compressed file
file_path = 'facebook_combined.txt.gz'

# Reading the graph edges from the file
with gzip.open(file_path, 'rt') as f:
    G = nx.read_edgelist(f, create_using=nx.Graph(), nodetype=int)
```

```python
# Print the basic information of the graph
print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
# Visualize the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=False, node_size=5)
plt.show()
```

```python
# Number of triangles
num_triangles = sum(nx.triangles(G).values()) // 3
print(f"Number of triangles: {num_triangles}")

# Check if the graph is connected and calculate the diameter
is_connected = nx.is_connected(G)
graph_diameter = nx.diameter(G) if is_connected else "Graph is not connected"
print(f"Is the graph connected? {is_connected}")
print(f"Diameter of the graph: {graph_diameter}")
```

```python
# Calculate the degree of all nodes
degrees = dict(G.degree())

# Find minimum and maximum degree
min_degree = min(degrees.values())
max_degree = max(degrees.values())

# Output the minimum and maximum degree
print(f"Minimum degree: {min_degree}")
print(f"Maximum degree: {max_degree}")

# Print the degrees of all nodes
print("Degrees of all nodes:", degrees)
```

```python
# Plotting the degree distribution
plt.figure(figsize=(8,6))
plt.hist(list(degrees.values()), bins=50, color='b', alpha=0.7)
plt.title('Degree Distribution')
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Plot betweenness centrality distribution
plt.figure(figsize=(8,6))
plt.hist(list(betweenness_centrality.values()), bins=50, color='g', alpha=0.7)
plt.title('Betweenness Centrality Distribution')
plt.xlabel('Betweenness Centrality')
plt.ylabel('Frequency')
plt.show()

# Print betweenness centrality for all vertices
print("Betweenness Centrality:", betweenness_centrality)
```

```python
# Calculate closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Plot closeness centrality distribution
plt.figure(figsize=(8,6))
plt.hist(list(closeness_centrality.values()), bins=50, color='r', alpha=0.7)
plt.title('Closeness Centrality Distribution')
plt.xlabel('Closeness Centrality')
plt.ylabel('Frequency')
plt.show()

# Print closeness centrality for all vertices
print("Closeness Centrality:", closeness_centrality)
```

```python
# Calculate clustering coefficient
clustering_coefficient = nx.clustering(G)

# Plot clustering coefficient distribution
plt.figure(figsize=(8,6))
plt.hist(list(clustering_coefficient.values()), bins=50, color='y', alpha=0.7)
plt.title('Clustering Coefficient Distribution')
plt.xlabel('Clustering Coefficient')
plt.ylabel('Frequency')
plt.show()

# Print clustering coefficient for all vertices
print("Clustering Coefficient:", clustering_coefficient)
```

```python
# Combined plot
plt.figure(figsize=(10,8))

# Degree Distribution
plt.subplot(2, 2, 1)
plt.hist(list(degrees.values()), bins=50, color='b', alpha=0.7)
plt.title('Degree Distribution')
plt.xlabel('Degree')

# Betweenness Centrality Distribution
plt.subplot(2, 2, 2)
plt.hist(list(betweenness_centrality.values()), bins=50, color='g', alpha=0.7)
plt.title('Betweenness Centrality Distribution')
plt.xlabel('Betweenness Centrality')

# Closeness Centrality Distribution
plt.subplot(2, 2, 3)
plt.hist(list(closeness_centrality.values()), bins=50, color='r', alpha=0.7)
plt.title('Closeness Centrality Distribution')
plt.xlabel('Closeness Centrality')

# Clustering Coefficient Distribution
plt.subplot(2, 2, 4)
plt.hist(list(clustering_coefficient.values()), bins=50, color='y', alpha=0.7)
plt.title('Clustering Coefficient Distribution')
plt.xlabel('Clustering Coefficient')

plt.tight_layout()
plt.show()
```
